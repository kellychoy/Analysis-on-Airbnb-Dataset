# import libraries

```python
import csv
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

# load the Airbnb listing dataset

```python
raw = pd.read_csv('listings_0903.csv')
```

```python
raw.sample(5)
```

```python
raw.describe()
```

```python
# filter out the unnecessary features
raw.drop(columns=['listing_url', 'scrape_id', 'last_scraped','picture_url','host_url','host_thumbnail_url', 'host_picture_url','calendar_last_scraped'], inplace=True)
```

```python
# further drop columns based on nan-check
# drop 'bathrooms', 'license', and 'calendar_updated', as nan account for >70% of the data
raw.drop(columns = ['bathrooms', 'license', 'calendar_updated','bathrooms_text'], inplace=True)
# drop review-related column
raw.drop(columns = ['first_review','last_review'], inplace=True)
# drop rows having null value for host_is_superhost
raw.dropna(subset=['host_is_superhost','host_has_profile_pic','host_name','host_since','host_response_time','host_acceptance_rate'], inplace = True)
```

```python
# select 'neighbourhood_cleansed' and drop the other 'neighbourhood_' columns
raw.drop(columns = ['neighborhood_overview', 'host_neighbourhood', 'neighbourhood','neighbourhood_group_cleansed','host_location'], inplace=True)
```

```python
raw.dropna(subset=['review_scores_rating','review_scores_accuracy','review_scores_cleanliness',
                   'review_scores_checkin','review_scores_communication','review_scores_location',
                   'review_scores_value','reviews_per_month'], inplace = True)
```

```python
raw.drop(columns=['latitude','longitude','bedrooms','beds'],inplace=True)
```

```python
# encoding: encode description, host_about into 1/0
raw['has_description'] = raw['description'].notna().astype(int)
raw['has_host_about'] = raw['host_about'].notna().astype(int)
```

```python
# encode amenities into cnt of amenities
raw['cnt_amenities'] = raw['amenities'].apply(len)

# encode host_verifications into cnt of verification
raw['cnt_verifications'] = raw['host_verifications'].apply(len)

# encode host_since into host for x years
raw['host_since'] = pd.to_datetime(raw['host_since'])
today = pd.Timestamp.today()
def month_difference(date1,date2):
    return (date2.year - date1.year) * 12 + date2.month - date1.month
raw['months_since'] = raw['host_since'].apply(lambda x: month_difference(x, today))

# encode price $35.00 to 35.00
raw['price'] = raw['price'].str.replace('$', '', regex=False).str.replace(',','',regex=False).astype(float)
```

```python
raw.drop(columns=['description','host_about','amenities','host_since','host_verifications'],inplace=True)
```

```python
# encode all 't/f' into 1/0
raw.replace({'t': 1, 'f': 0},inplace=True)
```

```python
# encode object to category
categorical_cols = ['host_response_time', 'neighbourhood_cleansed', 'property_type', 'room_type']
for col in categorical_cols:
    raw[col] = raw[col].astype('category')
# encode rate.str into float
raw['host_response_rate'] = raw['host_response_rate'].str.rstrip('%').astype('float') / 100.0
raw['host_acceptance_rate'] = raw['host_acceptance_rate'].str.rstrip('%').astype('float') / 100.0
```

```python
# check for null values
print(raw.isna().sum())
```

```python
# check the cleaned data
raw.describe()
```

```python
# save to csv
raw.to_csv('cleaned_listings.csv',index=False)
```
